## 20 Newsgroups dataset
- [20 Newsgroups data set](http://qwone.com/~jason/20Newsgroups/)
- [sklearn docs](https://scikit-learn.org/stable/datasets/real_world.html#newsgroups-dataset)

#### Import the libraries we'll use

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_20newsgroups
```

```python
# Load the "Twenty Newsgroups" dataset
dat = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'))

# Create a pandas DataFrame with the text and target variables
df = pd.DataFrame({'text': dat.data, 'target': dat.target})
```

```python
# Print the first few rows of the DataFrame to get an idea of the data
print(df.head(25))
```

```python
print("There are ", len(df.target.unique()), "classes.\n")
print(df.target.value_counts())
```

```python
# Visualize the distribution of samples in the target variable
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df['target'], bins=np.arange(21)-0.5, edgecolor='black')
ax.set_xticks(range(20))
ax.set_xticklabels(dat.target_names, rotation=90)
ax.set_xlabel('Target')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Samples in the Target Variable')
plt.show()
```

**Dataset:** 20 classes, 18846 samples, text features
**Objective:** Use text data to classify different news stories in different classes
**Algorithms:** Multi-class logistic regression -> decision trees/random forest -> gradient-boosted trees
